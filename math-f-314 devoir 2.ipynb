## 1. Introduction

La vitesse d'écoulement dans un milieu poreux est donnée par l'équation de Darcy

$$
\begin{equation}
\vec{u} = - \frac{K}{\mu} \left( \vec{\nabla} P - \rho g \vec{1_z} \right)
\end{equation}
$$

où $K$ est le coefficient de perméabilité du milieu, $ \mu $ est la viscosité du fluide et $g$ est la constante de gravité. Si un soluté est présent dans le fluide, sa concentration évolue comme

$$
\begin{equation}
\phi \frac{\partial C}{\partial t} = - \vec{u} \vec{\nabla}C + \phi D \vec{\nabla}^2 C
\end{equation}
$$

où $\phi$ est la porosité du millieu, $D$ est le coefficient de diffusion du soluté.

En manipulant ces équations, on arrive au système d'équation (1) et (2)

$$
\begin{equation}
\left( \frac{\partial^2}{\partial z^2} - k^2 \right) w = -k^2c \\
\frac{\partial c}{\partial t} - \frac{1}{Ra} \left( \frac{\partial ^2}{\partial z^2} - k^2 \right) c =  - \frac{\partial C_0}{\partial z} w
\end{equation}
$$

où $k$ est le nombre d'onde des perturbations, $Ra$ est le nombre de Rayleigh, $w$ est la composante verticale de la vitesse et $c$ est la concentration.
De plus, avec certaines conditions aux bords, on a l'équation

$$
\begin{equation}
\frac{\partial C_0}{\partial z} = - \frac{2}{\sqrt{\pi}} \sqrt{\frac{Ra}{4t}}e^{- \left( z \sqrt{\frac{Ra}{4t}} \right)^2}
\end{equation}
$$

### 2.1 Rappel du changement de variables

Soit un changement de variables réalisé à l'aide d'une bijection $f$

$$
\begin{equation}
x_1 = f_1(y_1,...,y_n) \\
\vdots \\
x_n = f_n(y_1,...,y_n)
\end{equation}
$$

Soit $g$ une fonction différentiable de laquelle on veut changer les variables tel que

$$
\begin{equation}
g(y_1,...,y_n) = g(x_1(y_1,...,y_n), ..., x_n(y_1,...,y_n))
\end{equation}
$$

Alors, la dérivée partielle en fonction de la variable $y_j$ peut s'écrire comme

$$
\begin{equation}
\frac{\partial g}{\partial y_j} = \sum_{i=1}^n \frac{\partial g}{\partial x_i} \frac{\partial x_i}{\partial y_j}
\end{equation}
$$


### 2.2 Solution

On a la nouvelle variable $\xi = z \sqrt{\frac{Ra}{4t}}$ et on cherche a trouver l'expression des dérivées des fonctions $w(\xi,t)$ et $c(\xi,t)$.
D'après le point 2.1, on a

$$
\frac{\partial f}{\partial z} = \frac{\partial f}{\partial \xi}\frac{\partial \xi}{\partial z}$$

Or on sait calculer
$$
\frac{\partial \xi}{\partial z} = \sqrt{\frac{Ra}{4t}} 
$$
Alors, on trouve
$$
\frac{\partial f}{\partial z} = \sqrt{\frac{Ra}{4t}} \frac{\partial f}{\partial \xi}
$$

On répète le même procédé pour trouver
$$
\frac{\partial^2 f}{\partial z^2} = \frac{Ra}{4t} \frac{\partial^2 f}{\partial \xi^2}
$$

En injectant ce résultat dans l'équation (1) et, on trouve

$$
\begin{equation}
\left( \frac{Ra}{4t} \frac{\partial^2 f}{\partial \xi^2} - k^2 \right) w = -k^2c
\end{equation}
$$

Pour l'équation (2), il faut aussi tenir compte de la dérivée en fonction du temps puisque la varable $\xi$ dépend de $t$. On a 

$$
\frac{\partial \ c(z,t)}{\partial t} = \frac{\partial \ c(\xi,t)}{\partial \xi}\frac{\partial \xi}{\partial t} + \frac{\partial c(\xi,t)}{\partial t}
$$

En injectant les résultats de $\frac{\partial^2 f}{\partial z^2}$ et $\frac{\partial c(z,t)}{\partial t}$ dans (2), on trouve

$$
\frac{\partial c(\xi,t)}{\partial t} - \frac{1}{t} \left( \frac{1}{4} \frac{\partial^2 }{\partial \xi^2} - \frac{\xi}{2} \frac{\partial}{\partial \xi} - \frac{k^2t}{Ra} \right) c = \sqrt{\frac{Ra}{\pi t}}e^{- \xi^2}w
$$

On trouve le système recherché
$$
\left( \frac{Ra}{4t} \frac{\partial^2 }{\partial \xi^2} - k^2 \right) w = -k^2c \\
\frac{\partial c(\xi,t)}{\partial t} - \frac{1}{t} \left( \frac{1}{4} \frac{\partial^2 }{\partial \xi^2} - \frac{\xi}{2} \frac{\partial}{\partial \xi} - \frac{k^2t}{Ra} \right) c = \sqrt{\frac{Ra}{\pi t}}e^{- \xi^2}w
$$

## 4. Question 3

On a $f(\xi) = e^{-\xi^2}h(\xi)$ tel que $h(\xi)$ est solution de l'équation de hermite $h'' - 2 \xi h' = -2 \lambda h$. On a l'opérateur $H$ défini comme $H = \frac{1}{4} \frac{\partial^2}{\partial \xi^2} + \frac{\xi}{2} \frac{\partial}{\partial \xi}$. On veut vérifier que $f$ est une fonction propre de $H$.

Avec ces définitions, on a

$$
\begin{align}
Hf &= \frac{1}{4} \frac{\partial}{\partial \xi} \left( -2\xi e^{-\xi^2}h(\xi) + e^{-\xi^2}h'(\xi) \right) + \frac{\xi}{2} \left( -2\xi e^{-\xi^2}h(\xi) + e^{-\xi^2}h'(\xi) \right) \\
   &= \frac{-1}{4} e^{-\xi^2} \left( h'' - 2\xi h' - 2h \right)
\end{align}
$$

où apparait l'équation de Hermite. Comme $h(\xi)$ est solution de cette équation, on peut écrire

$$
\begin{align}
Hf(\xi) &= \frac{1}{4} e^{-\xi^2} \left( -2 \lambda h - 2h \right) \\
        &= \frac{-1}{2} e^{-\xi^2} h ( \lambda + 1) \\
        &= -\frac{1}{2} ( \lambda + 1) f(\xi) \\
        &= \alpha f(\xi)
\end{align}
$$

Dès lors, il suit que 

$$
\alpha = -\frac{1}{2} ( \lambda + 1)
$$

On peut trouver des restrictions en utilisants les conditions aux bords $f(\xi = 0) = f(\xi = \infty) = 0$.
Les solutions de l'équation de Hermite sont les polynomes de Hermite. Donc pour chaque $\lambda$, $h(\xi)$ est un polynome de hermite.

$f(\xi = \infty) = 0$ implique que $\lambda$ est un entier positif.
$$
\lim_{\xi \to \infty}f(\xi) = e^{-\xi^2} h(\xi) = 0
$$
ce qui est toujours vrai puisqu'on a une exponentielle décroissante.

$f(\xi = 0) = 0$ implique que $\lambda$ est impair. En effet, en $\xi = 0$, on a $f(0) = h(0) = 0$. Or les polynomes de Hermite ne peuvent s'annuler en 0 que si leur indice est impair.

Les valaurs admissibles pour $\alpha$ sont donc $\alpha = \frac{-1}{2} ( \lambda + 1 )$ tel que $\lambda$ est un entier positif et impair.

On cherche a implémenter la matrice de l'opérateur $H$. L'implémentation de $H$ par la méthode des différences finie est très similaire à celle de l'opérateur $M_{i,j}$ réalisée dans la question 2. On trouve

$$
H_{i,j} =\frac{1}{4} \left(  \left( \frac{1}{\Delta^2} + i \right)u_{i+1} +  \left( \frac{1}{\Delta^2} - i \right)u_{i-1} -  \left( \frac{2}{\Delta^2} \right)u_i \right)
$$

ce qui correspond à la matrice

$$
\frac{1}{4}
\begin{pmatrix}
-\frac{2}{\Delta^2} & \frac{1}{\Delta^2} & 0 & 0 & 0 & 0 & 0 & 0 \\
\frac{1}{\Delta^2} -1 & -\frac{2}{\Delta^2} & \frac{1}{\Delta^2} +1 & 0 & 0 & 0 & 0 & 0  \\
0 & \frac{1}{\Delta^2} -2 & -\frac{2}{\Delta^2} & \frac{1}{\Delta^2} +2 & 0 & 0 & 0 & 0  \\
\vdots & 0 & \ddots & \ddots & \ddots\\
\\
0 & \ldots & & & 0 & \frac{1}{\Delta^2} - n +1 & -\frac{2}{\Delta^2} & \frac{1}{\Delta^2} + n -1 \\
0 & \ldots & & & & 0 & \frac{1}{\Delta^2} - n & -\frac{2}{\Delta^2} & 
\end{pmatrix}
$$

```python
import numpy as np #j'importe la bibliothèqye de numpy
```

maintenant je vais définir les variables données dans l'énoncé.

```python
l=5
n=101
delta=l/(n-1)
```

j'implémente la matrice $H_{i,j}$.

```python
h = np.zeros((n,n)) #Cette fonction crée une matrice carrée d'ordre n qui ne contient que des 0

#On modifie les éléments de matrices différents de 0.

for i in range(n):
    if i==0:
        h[i][i] = -2/(4*delta**2)
        h[i][i+1] = 1/(4*delta**2)+((i+1)/4) #je traite le première ligne séparément pour ne pas avoir un problème dans la boucle.
    if i==(n-1):
        h[i][i] = -2/(4*delta**2)
        h[i][i-1] = 1/(4*delta**2)-((i+1)/4) #je fais de même pour la dernière ligne de la matrice.
    else:
        h[i][i] = -2/(4*delta**2)
        h[i][i-1] = 1/(4*delta**2)-((i+1)/4)
        h[i][i+1] = 1/(4*delta**2)+((i+1)/4) #je remplace les valeurs restantes.
```

Jutilise la fonction linalg.eig de numpy pour calculer les valeurs propres de $H_{i,j}$.

```python
vp=np.linalg.eig(h)[0]
vp.sort()
for i in range(5):
    print(vp[len(vp)-1-i])
```

On retrouve bien les valeurs propres trouvées à la question 3.
